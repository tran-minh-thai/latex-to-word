```python
#@title Cell 2 - choose folders, then convert  { run: "auto" }
#@markdown Fill in the paths below (use the Drive file browser on the left to copy a path),
#@markdown then run this cell. The defaults point at the bundled `example/` paper.
import os
from google.colab import drive

# Mount Google Drive (no-op if already mounted)
if not os.path.exists('/content/drive/MyDrive'):
    drive.mount('/content/drive', force_remount=True)

LATEX_FOLDER  = "/content/drive/MyDrive/Researching/Latex2Word/example"          #@param {type:"string"}
WORD_TEMPLATE = "/content/drive/MyDrive/Researching/Latex2Word/reference.docx"   #@param {type:"string"}
MAIN_TEX      = "main.tex"                                                       #@param {type:"string"}
SCRIPT_PATH   = "/content/drive/MyDrive/Researching/Latex2Word/latex_to_word.py" #@param {type:"string"}
#@markdown HJS style: put the publication year in parentheses after the last author.
YEAR_AFTER_AUTHORS = True  #@param {type:"boolean"}

# The script copies WORD_TEMPLATE into LATEX_FOLDER as reference.docx, then runs
# pandoc. It also builds the bibliography (numbered by order of appearance, DOI
# per entry), resolves \ref/\eqref, and keeps pseudocode formulas as equations.
# NOTE: <MAIN_TEX>.bbl and <MAIN_TEX>.aux must exist (run LaTeX + BibTeX once).
if not os.path.exists(SCRIPT_PATH):
    print(f"ERROR: script not found: {SCRIPT_PATH}")
elif not os.path.isdir(LATEX_FOLDER):
    print(f"ERROR: folder not found: {LATEX_FOLDER}")
elif not os.path.exists(WORD_TEMPLATE):
    print(f"ERROR: template not found: {WORD_TEMPLATE}")
else:
    year_flag = "--year-after-authors" if YEAR_AFTER_AUTHORS else ""
    print(">>> Converting LaTeX -> Word ...")
    !python "{SCRIPT_PATH}" "{MAIN_TEX}" -d "{LATEX_FOLDER}" -t "{WORD_TEMPLATE}" {year_flag}
```

```python
# Cell 3 - download the Word file
import os
from google.colab import files

# Derive the output path from the choices made in Cell 2.
try:
    output_file = os.path.join(LATEX_FOLDER, os.path.splitext(MAIN_TEX)[0] + ".docx")
except NameError:
    output_file = "/content/drive/MyDrive/Researching/Latex2Word/example/main.docx"

if os.path.exists(output_file):
    files.download(output_file)
else:
    print(f"File not found: {output_file}")
```
